# Description des variables d'intêret 
* idmutinvar :Identifiant unique au niveau national permettant d'identifier chaque mutation présente dans la base de données. 
* coddep: code du département
* idnatmut : nature de mutation. 
* anneemut	annee de signature du document,
* moismut:	mois de signature du document,
* libnatmut :Libellé de nature de mutation. 
    Adjudication	Mutation onéreuse de type 'Adjudication' (idnatmut = 5)
    Echange	Mutation onéreuse de type 'Echange' (idnatmut = 6)
    Expropriation	Mutation onéreuse de type 'Expropriation' (idnatmut = 3)
    Vente	Mutation onéreuse de type 'Vente' (idnatmut = 1)
    Vente en l'état futur d'achèvement	Mutation onéreuse de type 'VEFA' (idnatmut = 2)
    Vente terrain à bâtir	Mutation onéreuse de type 'Vente de terrain à bâtir' (idnatmut = 4)
* valeurfonc: Valeur foncière en € telle que déclarée lors de la vente. 
    la valeur foncière correspond au prix sur lequel se sont mis d'accord les parties au moment de la mutation.
    Le prix indiqué correspond à tout ce qui est taxable, soit le prix net vendeur et inclut la TVA. 
    Ce prix n'inclut pas les frais de notaire, les frais d'agence ou la vente des biens meubles.
    *Des mutations présentes dans DVF ont des valeurs foncières très faibles voire nulles.
    Elles peuvent correspondre à des reventes entre filiales, entre collectivités et organisme HLM, etc.
* nbcomm :Nombre de communes sur lesquelles se situent l'ensemble des parcelles cadastrales concernées par la mutation.
* l_codinsee :Liste des codes INSEE des communes sur lesquelles se situent l'ensemble des parcelles cadastrales concernées par la mutation.
* sterr : Surface du terrain ayant muté qui correspond à la somme des surfaces des subdivisions fiscales (suf) composant la ou les parcelles (cf
* periodecst  : période de construction dominante des locaux ayant muté,
* nblocanc:	nombre de locaux anciens (année construction >= 5 ans)
* codtypbien : Typologie de biens proposée par le Groupe national DVF (GnDVF). Elle permet de classer chacune des mutations selon un ensemble défini de catégories de bien.

```python
def typebien(x):
    """
    Function to filter type of real state at 2 level according to GnDVF code
    1 Bati
    12 Appartement
    15 Bati mixt
    10 Indetermine
    11 Maison
    13 Depedance 
    14 Activité
    
    2 Non Bati
    20 Terrains ND
    21 Terrains TAB
    22 Terrain artificielle
    23 Terrain naturel
    
    Parameters
    ----------
    x : int
        Value from codtypbien columns.

    Returns
    -------
    2nd level Type du bien code.

    """
    strval = str(x)
    if strval[0] == "1":
        if strval[1] == "2":
            return 12
        elif strval[1] == "5":
            return 15
        elif strval[1] == "0":
            return 10
        elif strval[1] == "1":
            return 11
        elif strval[1] == "3":
            return 13
        elif strval[1] == "4":
            return 14
    if strval[0] == "2":
        if strval[1] == "0":
            return 20
        elif strval[1] == "1":
            return 21
        elif strval[1] == "2":
            return 22
        elif strval[1] == "3":
            return 23

def transfcolumn(x):
    # Converting string to list
    lx = ast.literal_eval(x)
    if len(lx) == 1:
        return str(lx[0])
    else:
        nlx=str(lx[0])
        for ix in lx[1:]:
            nlx = nlx+"_"+str(ix)
        return nlx
def valsupfB(row, valuecol):
    totalsupf = row["sterr"] + row["sbati"]
    if totalsupf>0:
        return row[valuecol]/totalsupf
    else:
        return np.nan
    
def stdfilter(row, valuemean,valuestd, alpha):
    """
    Apply outlier filter considering the standard deviation and mean of the population.
    valuemean : Mean value
    valuestd : Standard deviation
    alpha : threshold considered as in thresold = mean + alpha * std
    """
    if row >(valuemean + alpha*valuestd):
        return True
    else:
        return False
```

## Création de la variable cible *valsurf* 
Pour chaque mutation il existe une variable dans le set des données qui contient la surface vendue pour chaque type de bien (Bâtiment, Terrain et les sous-type). Parfois malgré la classification du type de biens, la surface peut être exprimé dans une autre catégorie, e.g. un bien décrit comme "Bâti" mais on trouve la surface vendue sous la categorie "Superficie Terrain"

```python
df1["valsurf"] = df1.apply(valsupfB,args=("valeurfonc",),axis=1)
```

```python
#Regression analysis
#Les communes seront "labellisés pour appliquer d'outil type PCA"
#Les mutations avec >1 communes auront une label unique pour chaque combinaison de commune
df1["communes"] = df1["l_codinsee"].apply(transfcolumn)
#For this case only dept 56
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df1["communes"].unique())
df1["commulbl"] = le.transform(df1["communes"])

#Si on veut regarder la difference. /!\ nbcomm >1 --> >>large superficie
dfsingle = df1[df1["nbcomm"]==1]
dfmulti = df1[df1["nbcomm"]>1]
#Pre selection des variables qui pourraient expliquer le prix
dfreg = df1[['valeurfonc', 'nbcomm', 'sterr', 'sbati', 'sbatmai',
       'sbatapt', 'sbatact', 'codtypbienlv1', 'valsurf',
       'totalsupf',"commulbl"]]
dfreg.dropna( how="any", axis=0,inplace=True)
```

```python
corr = dfreg.corr() #Create linear correlation matrix
```

```python
ax = sns.heatmap(corr, square=True,center=0,vmin=-.8,vmax=.8,cmap="RdBu_r")
```

```python
#K-cluster
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(dfreg)
kmeans = KMeans(n_clusters=10, random_state=0).fit(scaled_df)
print(len(kmeans.labels_))
```

```python
#PCL
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=dfreg.drop("valsurf",1)
y = dfreg["valsurf"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

pcr = make_pipeline(StandardScaler(), PCA(n_components=3), LinearRegression())
pcr.fit(X_train, y_train)
pca = pcr.named_steps['pca']
print(f"PCR r-squared with 4 components {pcr.score(X_test, y_test):.3f}")
#Multiple linear regression
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
axes[0].scatter(pca.transform(X_test)[:,0], y_test, alpha=.3, label='ground truth')
axes[0].scatter(pca.transform(X_test)[:,0], pcr.predict(X_test), alpha=.3,
                label='predictions')
axes[0].set(xlabel='Projected data onto first PCA component',
            ylabel='y', title='PCR / PCA')
axes[0].legend()
axes[1].scatter(pca.transform(X_test)[:,1], y_test, alpha=.3, label='ground truth')
axes[1].scatter(pca.transform(X_test)[:,1], pcr.predict(X_test), alpha=.3,
                label='predictions')
axes[1].set(xlabel='Projected data onto 2nd PCA component',
            ylabel='y', title='PCR / PCA')
axes[1].legend()
```
